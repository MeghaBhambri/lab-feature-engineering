Use appropriate methods to clean the columns GEOCODE2, WEALTH1, ADI, DMA,and MSA.

checking nans and value count

```python
def null_values_and_value_count(df: pd.DataFrame,col):
        value_count = df[col].value_counts()
        null_count = df[col].isna().sum()
        print("Null count for", col , "is", null_count)
        print("and value_count", col , "is" , value_count)  
    
    
```

```python
null_values_and_value_count(data,'GEOCODE2')
```

# geocode2 cleaning

clean the columns GEOCODE2

```python
#replacing nans with mode
# Find the mode value
mode_value = data['GEOCODE2'].mode()[0]

# Replace NaN values with the mode
data['GEOCODE2'] = data['GEOCODE2'].fillna(mode_value)
```

```python
data['GEOCODE2'].value_counts()
```

```python
data['GEOCODE2'].isna().sum()
```

```python
# checking unique VALUES WITH OTHER
data['GEOCODE2'].unique()
```

```python
# REPLACING GECODE2 BLANK VALUES WITH OTHER
data['GEOCODE2'] = data['GEOCODE2'].replace(' ', 'Not Assigned')
```

```python
#check mean
np.mean(numerical[numerical['WEALTH1'].notna()]['WEALTH1'])
```

```python
#numerical['AGE'] = numerical["AGE"].fillna(np.mean(numerical['AGE']))
fig, ax = plt.subplots(1,3,figsize=(16,5))
sns.histplot(numerical['WEALTH1'], ax = ax[0])
sns.histplot(numerical['WEALTH1'].fillna(np.mean(numerical[numerical['WEALTH1'].notna()]['WEALTH1'])), ax = ax[1])
sns.histplot(numerical['WEALTH1'].fillna(np.median(numerical[numerical['WEALTH1'].notna()]['WEALTH1'])),ax = ax[2])
plt.show()
numerical['WEALTH1'] = numerical['WEALTH1'].fillna(np.median(numerical['WEALTH1']))
```

```python
#replace by mean
mean_WEALTH1 = np.mean(numerical.loc[numerical['WEALTH1'].notna(), 'WEALTH1'])
numerical['WEALTH1'].fillna(mean_WEALTH1, inplace=True)
```
